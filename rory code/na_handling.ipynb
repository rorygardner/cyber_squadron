```python
import pandas as pd
import numpy as np


monthly = pd.read_csv("monthly.csv")
labs_df = pd.read_csv("labels_df.csv")

features = [
    "retail_and_recreation_percent_change_from_baseline", 
    "grocery_and_pharmacy_percent_change_from_baseline", 
    "parks_percent_change_from_baseline", 
    "transit_stations_percent_change_from_baseline", 
    "workplaces_percent_change_from_baseline", 
    "residential_percent_change_from_baseline"
]

county_mean = monthly.groupby("sub_region_2")[features].mean()

labels = labs_df.set_index("sub_region_2").loc[county_mean.index]["metro_label"]
print(county_mean.shape)
print(labels.value_counts())
```

```python
mobility = pd.read_csv("va_mobility.csv")
na_ratio = mobility.isna().mean()
print(na_ratio)
```

```python
county_missing = mobility.groupby('sub_region_2')[features].apply(lambda df: df.isna().sum(axis=1).sum())
county_total = mobility.groupby('sub_region_2')[features].size()*len(features)
county_missing_pct = (county_missing / county_total) * 100
county_missing_sorted = county_missing_pct.sort_values(ascending=False)

counties_keep = county_missing_pct[county_missing_pct < 50].index

mobility_filtered = mobility[mobility['sub_region_2'].isin(counties_keep)]
mobility_filtered.to_csv("mobility_filtered.csv", index=False)
```
